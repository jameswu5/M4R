```python
import torch
import torch.nn as nn
from trainer import NeuralNetworkTrainer
```

```python
class Parameters:
    def __init__(self, r, sigma, K, T):
        self.r = r
        self.sigma = sigma
        self.K = K
        self.T = T

r = 0.05
sigma = 0.2
K = 1.0
T = 1.0

params = Parameters(r=r, sigma=sigma, K=K, T=T)
```

```python
def payoff(S, K):
    return torch.relu(K - S)
```

```python
trainer = NeuralNetworkTrainer(params, payoff, 42)
trainer.train(10000);
```

```python
trainer.plot_losses()
```
