```python
import numpy as np
import matplotlib.pyplot as plt
```

## Call and put option payoffs

```python
K = 100

call = lambda x: np.maximum(x - K, 0)
put = lambda x: np.maximum(K - x, 0)
```

```python
S = np.linspace(85, 115, 100)

call_values = call(S)
put_values = put(S)
```

```python
plt.figure(figsize=(11, 5))

plt.subplot(1, 2, 1)
plt.plot(S, call_values)
plt.title('Call Option Payoff')
plt.xlabel('Stock Price at Maturity')
plt.ylabel('Payoff')

plt.subplot(1, 2, 2)
plt.plot(S, put_values)
plt.title('Put Option Payoff')
plt.xlabel('Stock Price at Maturity')
plt.ylabel('Payoff')

plt.tight_layout()
plt.show()
```
