```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

from heston.pinn.heston_pinn_nd import HestonMultiAssetPINN
from heston.tree import HestonTreeFast
from config.heston_nd import *
from config.plot_settings import *

from utility.continuation_prob import compute_continuation_probs, estimate_continuation_value_heston_nd, continuation_normal
```

## Load models

### PINNs

```python
seeds = [42]
pinns = []

for seed in seeds:
    pinn = HestonMultiAssetPINN(model_config, seed)
    pinn.set_params(K, r, T, kappa, theta, sigma_bar, sigmas, corr, rho_cross, S_min, S_max, V_min, V_max)
    pinns.append(pinn)
    pinn.load(f'../../models/heston_pinn_nd/{seed}.pth')
```

### Tree

#### One dimensional reduction

```python
A = np.sum(np.square(sigmas))
B = (1 - rho_asset) * A + rho_asset * np.square(np.sum(sigmas))

r_1d = n_assets * r + 0.5 * v0 * (B - A)
kappa_1d = kappa
theta_1d = B * theta
sigma_bar_1d = np.sqrt(B) * sigma_bar

n = 100
mv = 3 * n
mz = 6 * n
V0_min = v0 * 0.25
V0_max = v0 * 3
S0_min = S0 * 0.01
S0_max = S0 * 3

rho = np.mean(rho_cross)

heston_tree = HestonTreeFast(n, mz, mv, K, T, r_1d, kappa_1d, theta_1d, sigma_bar_1d, rho)
heston_tree.build_tree(V0_min=V0_min, V0_max=V0_max, S0_min=S0_min**2, S0_max=S0_max**2,
                       option_type="put", exercise_type="american")
```

## Comparing PINN with Tree

### Vary $S^1$, $S^2$, fix $V=0.04$ and $t=0$

```python
resolution = 100

V = v0
S1s = np.linspace(S0_min, S0_max, resolution)
S2s = np.linspace(S0_min, S0_max, resolution)
S1_grid, S2_grid = np.meshgrid(S1s, S2s)
S_grid = np.stack([S1_grid, S2_grid], axis=-1).reshape(-1, 2)
t = np.zeros((S_grid.shape[0], 1))  # Time to maturity is 0 at maturity

# PINN prices
nn_prices_all = np.zeros((resolution, resolution, len(pinns)))
S_tensor = torch.tensor(S_grid, dtype=torch.float32)
t_tensor = torch.tensor(t, dtype=torch.float32)
V_tensor = torch.tensor([[V]], dtype=torch.float32).repeat(S_tensor.shape[0], 1)

for i, pinn in enumerate(pinns):
    nn_prices_all[:, :, i] = pinn.predict(t_tensor, S_tensor, V_tensor).numpy().reshape(resolution, resolution)

nn_prices = nn_prices_all.mean(axis=-1)

# Heston tree prices
S_prod = S_grid[:, 0] * S_grid[:, 1]
ht_prices = heston_tree.price(V, S_prod, 0).reshape(resolution, resolution)
```

```python
# Plotting the results
Xs = np.linspace(K / S_max, S_max, 100)
Ys = K / Xs

plt.figure(figsize=figsize3)

# Plot the neural network price surface
plt.subplot(1, 3, 1)
vmax = np.abs(nn_prices).max()
vmin = -vmax
im = plt.imshow(nn_prices, extent=[S1s[0], S1s[-1], S2s[0], S2s[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=vmin, vmax=vmax)
plt.colorbar(im, label='NN price')
plt.plot(Xs, Ys, color=GRAPHITE, linestyle='--', label=r'$S^1 \times S^2 = K$')
plt.xlabel('Asset Price $S^1$')
plt.ylabel('Asset Price $S^2$')
plt.title('PINN price surface at t=0, V=V0')
plt.legend()

plt.subplot(1, 3, 2)
vmax = np.abs(ht_prices).max()
vmin = -vmax
im = plt.imshow(ht_prices, extent=[S1s[0], S1s[-1], S2s[0], S2s[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=vmin, vmax=vmax)
plt.colorbar(im, label='Heston Tree price')
plt.plot(Xs, Ys, color=GRAPHITE, linestyle='--', label=r'$S^1 \times S^2 = K$')
plt.xlabel('Asset Price $S^1$')
plt.ylabel('Asset Price $S^2$')
plt.title('Tree price surface at t=0, V=V0')
plt.legend()

plt.subplot(1, 3, 3)
diff = nn_prices - ht_prices

vmax = np.abs(diff).max()
vmin = -vmax
im = plt.imshow(diff, extent=[S1s[0], S1s[-1], S2s[0], S2s[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=vmin, vmax=vmax)
plt.colorbar(im, label='Difference (NN - Heston Tree)')
plt.plot(Xs, Ys, color=GRAPHITE, linestyle='--', label=r'$S^1 \times S^2 = K$')
plt.xlabel('Asset Price $S^1$')
plt.ylabel('Asset Price $S^2$')
plt.title('Price difference at t=0, V=V0')
plt.legend()

plt.tight_layout()
plt.show()
```

### Keep $S_1 = S_2$, vary $V$, fix $t=0$

```python
resolution = 100

Ss = np.linspace(S0_min, S0_max, resolution)
Vs = np.linspace(V0_min, V0_max, resolution)
S_grid_1d, V_grid = np.meshgrid(Ss, Vs)  # both shape (resolution, resolution)

S_pairs = np.stack([S_grid_1d.ravel(), S_grid_1d.ravel()], axis=-1)  # S1 = S2
V_flat = V_grid.ravel().reshape(-1, 1)
t = np.zeros((S_pairs.shape[0], 1))

# PINN prices
S_tensor = torch.tensor(S_pairs, dtype=torch.float32)
t_tensor = torch.tensor(t, dtype=torch.float32)
V_tensor = torch.tensor(V_flat, dtype=torch.float32)

nn_prices_all = np.zeros((resolution, resolution, len(pinns)))
for i, pinn in enumerate(pinns):
    nn_prices_all[:, :, i] = pinn.predict(t_tensor, S_tensor, V_tensor).numpy().reshape(resolution, resolution)

nn_prices_sv = nn_prices_all.mean(axis=-1)

# Heston tree prices
S_prod = S_pairs[:, 0] * S_pairs[:, 1]  # = S^2

ht_prices_sv = heston_tree.price(V_flat.ravel(), S_prod, 0).reshape(resolution, resolution)
```

```python
# Plotting
plt.figure(figsize=figsize3)

plt.subplot(1, 3, 1)
vmax = np.abs(nn_prices_sv).max()
im = plt.imshow(nn_prices_sv, extent=[Ss[0], Ss[-1], Vs[0], Vs[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='NN price')
plt.xlabel('Asset Price $S$ with $S^1 = S^2$')
plt.ylabel('Variance $V$')
plt.title(r'PINN price surface at $t=0$, $S^1=S^2$')

plt.subplot(1, 3, 2)
vmax = np.abs(ht_prices_sv).max()
im = plt.imshow(ht_prices_sv, extent=[Ss[0], Ss[-1], Vs[0], Vs[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='Heston Tree price')
plt.xlabel('Asset Price $S$ with $S^1 = S^2$')
plt.ylabel('Variance $V$')
plt.title(r'Tree price surface at $t=0$, $S^1=S^2$')

plt.subplot(1, 3, 3)
diff = nn_prices_sv - ht_prices_sv
vmax = np.abs(diff).max()
im = plt.imshow(diff, extent=[Ss[0], Ss[-1], Vs[0], Vs[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='Difference (NN - Heston Tree)')
plt.xlabel('Asset Price $S$ with $S^1 = S^2$')
plt.ylabel('Variance $V$')
plt.title(r'Price difference at $t=0$, $S^1=S^2$')

plt.tight_layout()
plt.show()
```

### Keep $S1 = S2 = \sqrt K$, vary $V$, vary $t$ 

```python
resolution = 100

S_atm = np.sqrt(K)  # S1 = S2 = sqrt(K), so S1*S2 = K
Vs = np.linspace(V0_min, V0_max, resolution)
ts = np.linspace(0, 1 - 1e-4, resolution)  # time to maturity

# meshgrid: rows = t index, cols = V index
V_grid, t_grid = np.meshgrid(Vs, ts)
V_flat = V_grid.ravel().reshape(-1, 1)
t_flat = t_grid.ravel().reshape(-1, 1)
S_flat = np.full((len(V_flat), 2), S_atm)

# PINN prices
S_tensor = torch.tensor(S_flat, dtype=torch.float32)
t_tensor = torch.tensor(t_flat, dtype=torch.float32)
V_tensor = torch.tensor(V_flat, dtype=torch.float32)

nn_prices_all = np.zeros((resolution, resolution, len(pinns)))
for i, pinn in enumerate(pinns):
    nn_prices_all[:, :, i] = pinn.predict(t_tensor, S_tensor, V_tensor).numpy().reshape(resolution, resolution)

nn_prices_vt = nn_prices_all.mean(axis=-1)

# Tree prices: batch over V at each time step k = round(t * n)
S_prod = S_atm ** 2  # = K
ht_prices_vt = np.zeros((resolution, resolution))
for j, t_val in enumerate(ts):
    k = int(round(t_val * n))
    k = min(k, n - 1)  # ensure k does not exceed n-1
    ht_prices_vt[j, :] = heston_tree.price(Vs, S_prod, k)
```

```python
# Plotting
plt.figure(figsize=figsize3)

plt.subplot(1, 3, 1)
vmax = np.abs(nn_prices_vt).max()
im = plt.imshow(nn_prices_vt, extent=[Vs[0], Vs[-1], ts[0], ts[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='NN price')
plt.xlabel('Variance $V$')
plt.ylabel('Time to maturity $t$')
plt.title(r'PINN price surface, $S_1=S_2=\sqrt{K}$')

plt.subplot(1, 3, 2)
vmax = np.abs(ht_prices_vt).max()
im = plt.imshow(ht_prices_vt, extent=[Vs[0], Vs[-1], ts[0], ts[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='Heston Tree price')
plt.xlabel('Variance $V$')
plt.ylabel('Time to maturity $t$')
plt.title(r'Tree price surface, $S_1=S_2=\sqrt{K}$')

plt.subplot(1, 3, 3)
diff = nn_prices_vt - ht_prices_vt
vmax = np.abs(diff).max()
im = plt.imshow(diff, extent=[Vs[0], Vs[-1], ts[0], ts[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='Difference (NN - Heston Tree)')
plt.xlabel('Variance $V$')
plt.ylabel('Time to maturity $t$')
plt.title(r'Price difference, $S_1=S_2=\sqrt{K}$')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=figsize3)

plt.subplot(1, 3, 1)
diff = nn_prices - ht_prices

vmax = np.abs(diff).max()
vmin = -vmax
im = plt.imshow(diff, extent=[S1s[0], S1s[-1], S2s[0], S2s[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=vmin, vmax=vmax)
plt.colorbar(im, label='Difference (NN - Heston Tree)')
plt.plot(Xs, Ys, color=GRAPHITE, linestyle='--', label=r'$S^1 \times S^2 = K$')
plt.xlabel(r'Asset Price $S^1$')
plt.ylabel(r'Asset Price $S^2$')
plt.title(r'Price difference at $t=0$, $V=V_0$')
plt.legend()

plt.subplot(1, 3, 2)
diff = nn_prices_sv - ht_prices_sv
vmax = np.abs(diff).max()
im = plt.imshow(diff, extent=[Ss[0], Ss[-1], Vs[0], Vs[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='Difference (NN - Heston Tree)')
plt.xlabel(r'Asset Price $S$ ($S^1 = S^2$)')
plt.ylabel(r'Variance $V$')
plt.title(r'Price difference at $t=0$, $S^1=S^2$')

plt.subplot(1, 3, 3)
diff = nn_prices_vt - ht_prices_vt
vmax = np.abs(diff).max()
im = plt.imshow(diff, extent=[Vs[0], Vs[-1], ts[0], ts[-1]],
                aspect='auto', origin='lower', cmap=cmap, vmin=-vmax, vmax=vmax)
plt.colorbar(im, label='Difference (NN - Heston Tree)')
plt.xlabel(r'Variance $V$')
plt.ylabel(r'Time to maturity $t$')
plt.title(r'Price difference with $S^1=S^2=\sqrt{K}$')

plt.tight_layout()
plt.savefig('../../plots/heston/heston_nd_comparison_heatmap.png')
plt.show()
```

## Continuation probabilities

### $S^1$-$S^2$ cross section

```python
resolution = 100
S1s = np.linspace(S0_min, S0_max, resolution)
S2s = np.linspace(S0_min, S0_max, resolution)
t = 0.0
V = v0

intrinsics = np.maximum(K - S1s[:, None] * S2s[None, :], 0) + 1e-9

nn_continuation_values_all = np.zeros((resolution, resolution, len(pinns)))
ht_continuation_values = np.zeros((resolution, resolution))

for i, S1 in enumerate(S1s):
    for j, S2 in enumerate(S2s):
        S = np.array([S1, S2])
        ht_continuation_values[i, j] = heston_tree.price(V, S1 * S2, 0, continuation_value=True)
        for k, pinn in enumerate(pinns):
            nn_continuation_values_all[i, j, k] = estimate_continuation_value_heston_nd(
                pinn.predict, t, S, V, r, kappa, theta, sigma_bar, sigmas, corr, rho_cross,
                n_paths=100, seed=42
            )
```

```python
nn_continuation_values = np.mean(nn_continuation_values_all, axis=2)
nn_continuation_values_std = np.std(nn_continuation_values_all, axis=2, ddof=1) * np.sqrt(len(pinns))

ht_continuation_probs = np.where(ht_continuation_values > intrinsics, 1.0, 0.0)
nn_continuation_probs_sigmoid = compute_continuation_probs(nn_continuation_values, intrinsics, eps=0.1, one=0.99, shift=0)

# Contour where S1 * S2 = K
Xs = np.linspace(K / S0_max, S0_max, resolution)
Ys = K / Xs

plt.figure(figsize=figsize2)

plt.subplot(1, 2, 1)
plt.imshow(ht_continuation_probs, extent=[S1s[0], S1s[-1], S2s[0], S2s[-1]],
           origin='lower', aspect='auto', cmap=cmap)
plt.plot(Xs, Ys, color='white', linestyle='--')
plt.colorbar(label='Continuation Probability')
plt.xlabel('Asset Price $S^1$')
plt.ylabel('Asset Price $S^2$')
plt.title('Tree Continuation Probabilities')

plt.subplot(1, 2, 2)
plt.imshow(nn_continuation_probs_sigmoid, extent=[S1s[0], S1s[-1], S2s[0], S2s[-1]],
           origin='lower', aspect='auto', cmap=cmap)
plt.plot(Xs, Ys, color='white', linestyle='--')
plt.colorbar(label='Continuation Probability')
plt.xlabel('Asset Price $S^1$')
plt.ylabel('Asset Price $S^2$')
plt.title('PINN Continuation Probabilities (Sigmoid)')

plt.tight_layout()
plt.savefig('../../plots/heston/heston_nd_continuation_probs.png')
plt.show()
```

### $S^1 = S^2$ diagonal, varying $t$

```python
ts = np.linspace(0, T - 1e-4, resolution)
S1s_time = np.linspace(0.8, np.sqrt(K), resolution)  # S1 = S2 = S, so S1*S2 = S^2

intrinsics_time = np.tile(np.maximum(K - S1s_time ** 2, 0) + 1e-9, (resolution, 1))  # shape (resolution, resolution)

nn_continuation_values_time_all = np.zeros((resolution, resolution, len(pinns)))
ht_continuation_values_time = np.zeros((resolution, resolution))

for i, t_val in enumerate(ts):
    k = min(int(round(t_val * n)), n - 1)
    ht_continuation_values_time[i, :] = heston_tree.price(v0, S1s_time ** 2, k, continuation_value=True)
    for j, S_val in enumerate(S1s_time):
        S = np.array([S_val, S_val])
        for m, pinn in enumerate(pinns):
            nn_continuation_values_time_all[i, j, m] = estimate_continuation_value_heston_nd(
                pinn.predict, t_val, S, v0, r, kappa, theta, sigma_bar, sigmas, corr, rho_cross,
                n_paths=100, seed=42
            )
```

```python
nn_continuation_values_time = np.mean(nn_continuation_values_time_all, axis=2)
nn_continuation_values_time_std = np.std(nn_continuation_values_time_all, axis=2, ddof=1) * np.sqrt(len(pinns))

ht_continuation_probs_time = np.where(ht_continuation_values_time > intrinsics_time, 1.0, 0.0)
nn_continuation_probs_time_sigmoid = compute_continuation_probs(nn_continuation_values_time, intrinsics_time, eps=0.1, one=0.99, shift=0)
# nn_continuation_probs_time_gaussian = continuation_normal(nn_continuation_values_time, nn_continuation_values_time_std, intrinsics_time)

plt.figure(figsize=figsize2)

plt.subplot(1, 2, 1)
plt.imshow(ht_continuation_probs_time, extent=[S1s_time[0], S1s_time[-1], ts[0], ts[-1]],
           origin='lower', aspect='auto', cmap=cmap)
plt.colorbar(label='Continuation Probability')
plt.xlabel(r'Asset Price $S$ ($S^1 = S^2 = S$)')
plt.ylabel('Elapsed time $t$')
plt.title('Tree Continuation Probabilities')

plt.subplot(1, 2, 2)
plt.imshow(nn_continuation_probs_time_sigmoid, extent=[S1s_time[0], S1s_time[-1], ts[0], ts[-1]],
           origin='lower', aspect='auto', cmap=cmap)
plt.colorbar(label='Continuation Probability')
plt.xlabel(r'Asset Price $S$ ($S^1 = S^2 = S$)')
plt.ylabel('Elapsed time $t$')
plt.title('PINN Continuation Probabilities (Sigmoid)')

plt.tight_layout()
plt.savefig('../../plots/heston/heston_nd_continuation_probs_time.png')
plt.show()
```
